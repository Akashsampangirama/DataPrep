```python
def listToString(s):  
    
    # initialize an empty string 
    str1 = " "  
    
    # traverse in the string   
    for ele in s: 
        ele = f'{ele}'+" "
        str1 += ele 
    
    # return string   
    return str1

class my_dictionary(dict): 
  
    # __init__ function 
    def __init__(self): 
        self = dict() 
          
    # Function to add key:value 
    def add(self, key, value): 
        self[key] = value 

import difflib
def closeMatches(patterns, word): 
     return difflib.get_close_matches(word, patterns)
```

```python
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize
from jellyfish import soundex
```

```python
import re
dict_obj_upc = my_dictionary()
dict_obj_tokenOld = my_dictionary()
dict_obj_tokenNew = my_dictionary()
dict_obj_tkn = my_dictionary()
dict_obj_tknold = my_dictionary()
dict_obj_cntry = my_dictionary()
allinone = []
#for upc in df_new_base['old_upc14'][(df_new_base['old_upc14'].isin(df_change_cnt['upc14']))==False]:#df_new_base['old_upc14']:#['10000314258','10000341569','10000325019','97018148892847']:
for upc in df_analyse.index:#['10000314258','10000341569','10000325019','97018148892847']:
    #print(upc)
    dict_obj_old = my_dictionary()
    dict_obj_new = my_dictionary()
    dict_obj_diff = my_dictionary()
    #.tolist() listToString()
    s = re.sub(r"[^a-z.A-Z0-9]+", ' ',df_analyse['DESC_BANK_ITEM_OLD'][upc])
    k = re.sub(r"[^a-z.A-Z0-9]+", ' ',df_analyse['DESC_BANK_ITEM_NAME'][upc])
    ln_cntry = df_analyse['COUNTRY_ID_OLD'][upc].item()
    #print(ln_cntry)
    X_list_old = word_tokenize(s)
    X_list_new = word_tokenize(k)

    ln_filtr_words = []
    ln_local_pattern = X_list_old
    for j in ln_local_pattern:
        ln_local = closeMatches(ln_local_pattern,j)
        if len(ln_local)>0:
            ln_filtr_words.append(ln_local[0])
        ln_local_pattern = [word for word in ln_local_pattern if word not in ln_local]
    X_list_old = ln_filtr_words
    for i in X_list_old:
        dict_obj_old.add(soundex(i),i)

    ln_filtr_words = []
    ln_local_pattern = X_list_new
    for j in ln_local_pattern:
        ln_local = closeMatches(ln_local_pattern,j)
        if len(ln_local)>0:
            ln_filtr_words.append(ln_local[0])
        ln_local_pattern = [word for word in ln_local_pattern if word not in ln_local]
    X_list_new = ln_filtr_words
    for i in X_list_new:
        dict_obj_new.add(soundex(i),i)
        
    ln_tokenOld = ''
    ln_tokenNew = ''
    for key in dict_obj_old.keys():
        if not key in dict_obj_new:
            dict_obj_diff.add(key,1)
            ln_tokenOld = ln_tokenOld + ' ' + dict_obj_old.get(key)
    for key in dict_obj_new.keys():
        if not key in dict_obj_old:
            ln_tokenNew = ln_tokenNew + ' ' + dict_obj_new.get(key)
            dict_obj_diff.add(key,1)
    allinone.append([df_analyse['SVI'][upc],len(word_tokenize(ln_tokenNew)),len(word_tokenize(ln_tokenOld)),ln_tokenNew,ln_tokenOld,len(dict_obj_new),ln_cntry])
```
